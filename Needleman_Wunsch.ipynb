```
import numpy as np
from string import *
import pandas as pd 

#Declare the dictionary for BLOSUM62 Score
blosum62 = {
    ('W', 'F'): 1, ('L', 'R'): -2, ('S', 'P'): -1, ('V', 'T'): 0,
    ('Q', 'Q'): 5, ('N', 'A'): -2, ('Z', 'Y'): -2, ('W', 'R'): -3,
    ('Q', 'A'): -1, ('S', 'D'): 0, ('H', 'H'): 8, ('S', 'H'): -1,
    ('H', 'D'): -1, ('L', 'N'): -3, ('W', 'A'): -3, ('Y', 'M'): -1,
    ('G', 'R'): -2, ('Y', 'I'): -1, ('Y', 'E'): -2, ('B', 'Y'): -3,
    ('Y', 'A'): -2, ('V', 'D'): -3, ('B', 'S'): 0, ('Y', 'Y'): 7,
    ('G', 'N'): 0, ('E', 'C'): -4, ('Y', 'Q'): -1, ('Z', 'Z'): 4,
    ('V', 'A'): 0, ('C', 'C'): 9, ('M', 'R'): -1, ('V', 'E'): -2,
    ('T', 'N'): 0, ('P', 'P'): 7, ('V', 'I'): 3, ('V', 'S'): -2,
    ('Z', 'P'): -1, ('V', 'M'): 1, ('T', 'F'): -2, ('V', 'Q'): -2,
    ('K', 'K'): 5, ('P', 'D'): -1, ('I', 'H'): -3, ('I', 'D'): -3,
    ('T', 'R'): -1, ('P', 'L'): -3, ('K', 'G'): -2, ('M', 'N'): -2,
    ('P', 'H'): -2, ('F', 'Q'): -3, ('Z', 'G'): -2, ('X', 'L'): -1,
    ('T', 'M'): -1, ('Z', 'C'): -3, ('X', 'H'): -1, ('D', 'R'): -2,
    ('B', 'W'): -4, ('X', 'D'): -1, ('Z', 'K'): 1, ('F', 'A'): -2,
    ('Z', 'W'): -3, ('F', 'E'): -3, ('D', 'N'): 1, ('B', 'K'): 0,
    ('X', 'X'): -1, ('F', 'I'): 0, ('B', 'G'): -1, ('X', 'T'): 0,
    ('F', 'M'): 0, ('B', 'C'): -3, ('Z', 'I'): -3, ('Z', 'V'): -2,
    ('S', 'S'): 4, ('L', 'Q'): -2, ('W', 'E'): -3, ('Q', 'R'): 1,
    ('N', 'N'): 6, ('W', 'M'): -1, ('Q', 'C'): -3, ('W', 'I'): -3,
    ('S', 'C'): -1, ('L', 'A'): -1, ('S', 'G'): 0, ('L', 'E'): -3,
    ('W', 'Q'): -2, ('H', 'G'): -2, ('S', 'K'): 0, ('Q', 'N'): 0,
    ('N', 'R'): 0, ('H', 'C'): -3, ('Y', 'N'): -2, ('G', 'Q'): -2,
    ('Y', 'F'): 3, ('C', 'A'): 0, ('V', 'L'): 1, ('G', 'E'): -2,
    ('G', 'A'): 0, ('K', 'R'): 2, ('E', 'D'): 2, ('Y', 'R'): -2,
    ('M', 'Q'): 0, ('T', 'I'): -1, ('C', 'D'): -3, ('V', 'F'): -1,
    ('T', 'A'): 0, ('T', 'P'): -1, ('B', 'P'): -2, ('T', 'E'): -1,
    ('V', 'N'): -3, ('P', 'G'): -2, ('M', 'A'): -1, ('K', 'H'): -1,
    ('V', 'R'): -3, ('P', 'C'): -3, ('M', 'E'): -2, ('K', 'L'): -2,
    ('V', 'V'): 4, ('M', 'I'): 1, ('T', 'Q'): -1, ('I', 'G'): -4,
    ('P', 'K'): -1, ('M', 'M'): 5, ('K', 'D'): -1, ('I', 'C'): -1,
    ('Z', 'D'): 1, ('F', 'R'): -3, ('X', 'K'): -1, ('Q', 'D'): 0,
    ('X', 'G'): -1, ('Z', 'L'): -3, ('X', 'C'): -2, ('Z', 'H'): 0,
    ('B', 'L'): -4, ('B', 'H'): 0, ('F', 'F'): 6, ('X', 'W'): -2,
    ('B', 'D'): 4, ('D', 'A'): -2, ('S', 'L'): -2, ('X', 'S'): 0,
    ('F', 'N'): -3, ('S', 'R'): -1, ('W', 'D'): -4, ('V', 'Y'): -1,
    ('W', 'L'): -2, ('H', 'R'): 0, ('W', 'H'): -2, ('H', 'N'): 1,
    ('W', 'T'): -2, ('T', 'T'): 5, ('S', 'F'): -2, ('W', 'P'): -4,
    ('L', 'D'): -4, ('B', 'I'): -3, ('L', 'H'): -3, ('S', 'N'): 1,
    ('B', 'T'): -1, ('L', 'L'): 4, ('Y', 'K'): -2, ('E', 'Q'): 2,
    ('Y', 'G'): -3, ('Z', 'S'): 0, ('Y', 'C'): -2, ('G', 'D'): -1,
    ('B', 'V'): -3, ('E', 'A'): -1, ('Y', 'W'): 2, ('E', 'E'): 5,
    ('Y', 'S'): -2, ('C', 'N'): -3, ('V', 'C'): -1, ('T', 'H'): -2,
    ('P', 'R'): -2, ('V', 'G'): -3, ('T', 'L'): -1, ('V', 'K'): -2,
    ('K', 'Q'): 1, ('R', 'A'): -1, ('I', 'R'): -3, ('T', 'D'): -1,
    ('P', 'F'): -4, ('I', 'N'): -3, ('K', 'I'): -3, ('M', 'D'): -3,
    ('V', 'W'): -3, ('W', 'W'): 11, ('M', 'H'): -2, ('P', 'N'): -2,
    ('K', 'A'): -1, ('M', 'L'): 2, ('K', 'E'): 1, ('Z', 'E'): 4,
    ('X', 'N'): -1, ('Z', 'A'): -1, ('Z', 'M'): -1, ('X', 'F'): -1,
    ('K', 'C'): -3, ('B', 'Q'): 0, ('X', 'B'): -1, ('B', 'M'): -3,
    ('F', 'C'): -2, ('Z', 'Q'): 3, ('X', 'Z'): -1, ('F', 'G'): -3,
    ('B', 'E'): 1, ('X', 'V'): -1, ('F', 'K'): -3, ('B', 'A'): -2,
    ('X', 'R'): -1, ('D', 'D'): 6, ('W', 'G'): -2, ('Z', 'F'): -3,
    ('S', 'Q'): 0, ('W', 'C'): -2, ('W', 'K'): -3, ('H', 'Q'): 0,
    ('L', 'C'): -1, ('W', 'N'): -4, ('S', 'A'): 1, ('L', 'G'): -4,
    ('W', 'S'): -3, ('S', 'E'): 0, ('H', 'E'): 0, ('S', 'I'): -2,
    ('H', 'A'): -2, ('S', 'M'): -1, ('Y', 'L'): -1, ('Y', 'H'): 2,
    ('Y', 'D'): -3, ('E', 'R'): 0, ('X', 'P'): -2, ('G', 'G'): 6,
    ('G', 'C'): -3, ('E', 'N'): 0, ('Y', 'T'): -2, ('Y', 'P'): -3,
    ('T', 'K'): -1, ('A', 'A'): 4, ('P', 'Q'): -1, ('T', 'C'): -1,
    ('V', 'H'): -3, ('T', 'G'): -2, ('I', 'Q'): -3, ('Z', 'T'): -1,
    ('C', 'R'): -3, ('V', 'P'): -2, ('P', 'E'): -1, ('M', 'C'): -1,
    ('K', 'N'): 0, ('I', 'I'): 4, ('P', 'A'): -1, ('M', 'G'): -3,
    ('T', 'S'): 1, ('I', 'E'): -3, ('P', 'M'): -2, ('M', 'K'): -1,
    ('I', 'A'): -1, ('P', 'I'): -3, ('R', 'R'): 5, ('X', 'M'): -1,
    ('L', 'I'): 2, ('X', 'I'): -1, ('Z', 'B'): 1, ('X', 'E'): -1,
    ('Z', 'N'): 0, ('X', 'A'): 0, ('B', 'R'): -1, ('B', 'N'): 3,
    ('F', 'D'): -3, ('X', 'Y'): -1, ('Z', 'R'): 0, ('F', 'H'): -1,
    ('B', 'F'): -3, ('F', 'L'): 0, ('X', 'Q'): -1, ('B', 'B'): 4
    }

#Function to calculate the score
def score_pairwise(seq1, seq2, matrix, gap_penalty):
  score = 0
  gap = False
  for i in range(len(seq1)):
        pair = (seq1[i], seq2[i])
        if not gap:
            if '-' in pair:
                gap = True
                score += gap_penalty
            else:
                score += score_match(pair, matrix)
        else:
            if '-' not in pair:
                gap = False
                score += score_match(pair, matrix)
            else:
                score += gap_penalty
  return score

#Function for matching the pair and show its score based on the dictionary
def score_match(pair, matrix):
    if pair not in matrix:
        return matrix[(tuple(reversed(pair)))]
    else:
        return matrix[pair]

#Function to decide the path
def Pointers(diag,hori,vert):
    pointer = max(diag,hori,vert) 

    if (diag == pointer):
        return 'D'
    elif (hori == pointer):
        return 'H'
    elif (vert == pointer):
        return 'V'    

#The main function of Needleman Wunsch
def Needleman_Wunsch(s1, s2, penalty):
    n = len(s1) + 1   #The dimension of matrix columns
    m = len(s2) + 1   #The dimension of matrix rows
    #Initialize all the matrices with 0
    align_matrix = np.zeros((m,n),dtype = int) 
    path_matrix = np.zeros((m,n),dtype = str) 
    blosum_matrix = np.zeros((m,n),dtype = int) 
    
    #Fill all the rows of the matrices with penalty
    for i in range(m):
        align_matrix[i][0] = penalty * i
        path_matrix[i][0] = 'V'
        blosum_matrix[i][0] = penalty * i
    #Fill all the column of the matrices with penalty
    for j in range (n):
        align_matrix[0][j] = penalty * j
        path_matrix [0][j] = 'H'
        blosum_matrix[0][j] = penalty * j


    path_matrix [0][0] = 0 
    for i in range(1,m):
        for j in range(1,n):
            diagonal   = align_matrix[i-1][j-1] + score_pairwise(s1[j-1],s2[i-1], blosum62, penalty)  #The value of the cell from diagonal direction
            horizontal = align_matrix[i][j-1] + penalty                                               #The value of the cell from horizontal direction
            vertical   = align_matrix[i-1][j] + penalty                                               #The value of the cell from vertical direction
            align_matrix[i][j] = max(diagonal,horizontal,vertical)                    #Fill the matrix with maximum value from the above calculation
            path_matrix[i][j] = Pointers(diagonal, horizontal, vertical)              #Fill the path of each cell from Pointers() function
            blosum_matrix[i][j] = score_pairwise(s1[j-1],s2[i-1], blosum62, penalty)  #Filling the BLOSUM62 score for each pair in cell

    #To print all the matrices
    sequ1 = seq1[:0] + ' ' + seq1[0:]   #Add blank space in the first position on sequence 1
    sequ2 = seq2[:0] + ' ' + seq2[0:]   #Add blank space in the first position on sequence 2
    #To show the matrices with the sequence
    b_mat = pd.DataFrame(blosum_matrix, index = list(sequ2), columns = list(sequ1)) 
    a_mat = pd.DataFrame(align_matrix, index = list(sequ2), columns = list(sequ1)) 
    p_mat = pd.DataFrame(path_matrix, index = list(sequ2), columns = list(sequ1))
    print('Blosum Table: ')
    print(b_mat)
    print('\nTable Score: ')
    print(a_mat)  
    print('\nTable Path: ')
    print(p_mat)

    #Getting the max value and its index of the last row in the matrix
    last_row = align_matrix[m-1]
    max_last_row = max(last_row)
    row_index = np.argwhere(align_matrix == max_last_row)
    x = row_index[[0][0]]
    x1 = x[0]
    x2 = x[1]
    #Getting the max value and its index of the last column in the matrix
    last_column = align_matrix[:, n-1]
    max_last_column = max(last_column)
    column_index = np.argwhere(align_matrix == max_last_column)
    y = column_index[[0][0]]
    y1 = y[0]
    y2 = y[1]

    #To make the alignment of the sequences, with the highest score, from the right bottom of the matrix 
    i, j = m-1, n-1
    while i > 0 and j > 0:
      if (max_last_row == max_last_column):      #If max value on both last row and last column are same
        if (x1 == y1 and x2 == y2):              #If the position of the max value is the same on both last  row and column
          point = path_matrix[i][j]
          if (point == 'D'):                     #If the pointer meet D, then go diagonal
            i = i-1
            j = j-1
          elif (point == 'H'):                   #If the pointer meet H, then go horizontal
            j = j-1
            s2 = s2[:i] + '-' + s2[i:]           #Making a gap in the sequence
          elif (point == 'V'):                   #If the pointer meet V, then go vertical
            i = i-1
            s1 = s1[:j] + '-' + s1[j:]
        else:                                    #If the positions are different
          point_mat1 = path_matrix[x1][x2]
          point_mat2 = path_matrix[y1][y2]
          if (point_mat1 == 'D'):
            i = i-1
            j = j-1
          elif (point_mat2 == 'D'):
            i = i-1
            j = j-1
          elif (point_mat1 == 'H'):
            j = j-1
            s2 = s2[:i] + '-' + s2[i:]
          elif (point_mat2 == 'H'):
            j = j-1
            s2 = s2[:i] + '-' + s2[i:]
          elif (point_mat1 == 'V'):
            i = i-1
            s1 = s1[:j] + '-' + s1[j:]
          elif (point_mat2 == 'V'):
            i = i-1
            s1 = s1[:j] + '-' + s1[j:]
      elif (max_last_row > max_last_column):    #If the max value is on the last row
        point1 = path_matrix[x1][x2]
        if (point1 == 'D'):
            i = i-1
            j = j-1
        elif (point1 == 'H'):
            j = j-1
            s2 = s2[:i] + '-' + s2[i:]
        elif (point1 == 'V'):
            i = i-1
            s1 = s1[:j] + '-' + s1[j:]
      elif (max_last_row < max_last_column):    #If the max value is on the last column
        point2 = path_matrix[x1][x2]
        if (point1 == 'D'):
            i = i-1
            j = j-1
        elif (point2 == 'H'):
            j = j-1
            s2 = s2[:i] + '-' + s2[i:]
        elif (point2 == 'V'):
            i = i-1
            s1 = s1[:j] + '-' + s1[j:]

    print('\nAlignment 1: ', s1)
    print('Alignment 2: ', s2)
    print('BLOSUM62 Score: ', score_pairwise(s1, s2, blosum62, penalty))

seq1 = 'MHHNALQRRTVWVNAY'
seq2 = 'MHHALQRRTVWVNAY'

Needleman_Wunsch(seq1, seq2, -6)
```
